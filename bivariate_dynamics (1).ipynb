### Stable Node

```python
""" 
Functions will all be in the form of f(Xt+1) = f(Xt) - define
function by using Xt.

Use numpy for math notation. E.g. np.sqrt(Xt)

"""

def f(Xt, Yt):
    return 0.6*Xt - 0.2*Yt

def g(Xt, Yt):
    return 0.2*Xt + 0.1*Yt


# Instantiating Dynamics object with a list of functions and initial_conditions.

initial_conditions = [.5, .5]
functions = [f, g]

dynamics = Dynamics(functions, 
                    initial_conditions
                    )

dynamics.plot_range(start = -1.0,
                    stop = 1.0,
                    step = 0.1,
                    figsize = (15,15),
                    divergent_length = 10,
                    accuracy = 0.001,
                    mode = 'bivariate'
                   )
```

### Unstable Node

```python
""" 
Functions will all be in the form of f(Xt+1) = f(Xt) - define
function by using Xt.

Use numpy for math notation. E.g. np.sqrt(Xt)

"""

def f(Xt, Yt):
    return 4*Xt - 2*Yt

def g(Xt, Yt):
    return -5*Xt + 5*Yt

# Instantiating Dynamics object with a list of functions and initial_conditions.

initial_conditions = [.5, .5]
functions = [f, g]

dynamics = Dynamics(functions, 
                    initial_conditions
                    )

dynamics.plot_range(start = -5.0,
                    stop = 5.0,
                    step = 0.5,
                    figsize = (15,15),
                    divergent_length = 1,
                    accuracy = 0.001,
                    mode = 'bivariate'
                   )
```

### Simple Migration Dynamics

```python
""" 
Functions will all be in the form of f(Xt+1) = f(Xt) - define
function by using Xt.

Use numpy for math notation. E.g. np.sqrt(Xt)

"""

def f(Xt, Yt):
    return 0.9*Xt + 0.05*Yt

def g(Xt, Yt):
    return -0.22*Xt + 0.61*Yt +0.25

# Instantiating Dynamics object with a list of functions and initial_conditions.

initial_conditions = [1, 1]
functions = [f, g]

dynamics = Dynamics(functions, 
                    initial_conditions
                    )

dynamics.plot_range(start = -1.0,
                    stop = 1.0,
                    step = 0.1,
                    figsize = (15,15),
                    divergent_length = 100,
                    accuracy = 0.0,
                    mode = 'bivariate'
                   )
```

```python
""" 
Functions will all be in the form of f(Xt+1) = f(Xt) - define
function by using Xt.

Use numpy for math notation. E.g. np.sqrt(Xt)

"""

def f(x, y):
    return np.sin(x)

def g(x, y):
    return -0.22*x + 0.61*y +0.25

# Instantiating Dynamics object with a list of functions and initial_conditions.

initial_conditions = [1,1]
functions = [f, g]

dynamics = Dynamics(functions, 
                    initial_conditions
                    )

dynamics.recurse(divergent_length=100,
                 accuracy = 0.1
                )

#dynamics.bivariate_plot()


dynamics.plot_range(start = -1.0,
                    stop = 1.0,
                    step = 0.1,
                    figsize = (15,15),
                    divergent_length = 10,
                    accuracy = 0.0,
                    mode = 'bivariate'
                   )
```
