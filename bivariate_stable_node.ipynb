```python
from Dynamics import bivariate_dynamics, bivariate_plot
```

```python
""" 
Functions will all be in the form of Xt+1 = x(Xt, Yt) - define
function by using Xt and Yt

Use numpy for math notation. E.g. np.sqrt(Xt, Yt)

"""
def f(Xt, Yt):
    return -0.4*Xt - 0.2*Yt

def g(Xt, Yt):
    return 0.2*Xt - 0.9*Yt
```

```python
# setting our initial condition and calling the dynamics function

x_init = .5
y_init = 0.5

x, y, x_path, y_path = bivariate_dynamics(f, g, x_init, y_init)
```

```python
# x & y convergence approximation

x, y
```
