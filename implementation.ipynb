```python
from Dynamics import Dynamics
from polycalc import Polynomial
```

```python
''' CHAOS '''

a = Polynomial([0,3.8,-2.91])
dynamics = Dynamics(functions = [a.evaluate], initial_conditions = [.9])
dynamics.model(divergent_length = 50, accuracy = 0.00001)
dynamics.univariate_plot(figsize = (12,12))
```

```python
''' Convergent Interval '''

a = Polynomial([0,0,1])

dynamics = Dynamics(functions = [a.evaluate], initial_conditions = [.9])
dynamics.model(divergent_length = 50, accuracy = 0.00001)
dynamics.univariate_plot(figsize = (12,12))
```

```python
''' Divergent Interval '''

dynamics = Dynamics(functions = [a.evaluate], initial_conditions = [1.1])
dynamics.model(divergent_length = 5, accuracy = 0.00001)
dynamics.univariate_plot(figsize = (12,12))
```
