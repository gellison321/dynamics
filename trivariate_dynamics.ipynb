### Simple Immigration Dynamics

```python
""" 
Functions will all be in the form of f(Xt+1) = f(Xt) - define
function by using Xt.

Use numpy for math notation. E.g. np.sqrt(Xt)

"""

def f(Xt, Yt, Zt):
    return 0.9*Xt + 0.05*Yt

def g(Xt, Yt, Zt):
    return 0.86*Yt +0.25*Zt +0.03*Xt

def h(Xt, Yt, Zt):
    return 0.75*Zt +0.09*Yt + 0.07*Xt
```

```python
# Instantiating Dynamics object with a list of functions and initial_conditions.

initial_conditions = [.5, .5, .5]
functions = [f, g, h]

dynamics = Dynamics(functions, 
                    initial_conditions
                    )


# Calling the model() function to model the dynamics until we've converged our diverged.


dynamics.recurse(divergent_length=100, # minimum recursions until the loop breaks
                 accuracy = 0.00001 # for convergence
                )

dynamics.trivariate_plot(figsize = (12,12))
```

```python
# plotting a matrix of initial_conditions

start = -1
stop = 1
step = 0.2


fig = plt.figure(figsize=(15,15))

ax = plt.axes(projection='3d');
ax.set_title('Phase Portrait');


for x in np.arange(start, stop, step):
    for y in np.arange(start, stop, step):
        for z in np.arange(start, stop, step):
            initial_conditions = [x, y, z]
            functions = [f, g, h]

            dynamics = Dynamics(functions, 
                                initial_conditions
                                )


            # Calling the model() function to model the dynamics until we've convergede or diverged.


            dynamics.recurse(divergent_length=100, # minimum recursions until the loop breaks
                             accuracy = 0.00001 # for convergence
                            )

            x_path = dynamics.X[0]
            y_path = dynamics.X[1]
            z_path = dynamics.X[2]
            
            ax.plot([x], [y], [z], marker = "o", markersize = 2, color = "red")

            ax.plot3D(x_path, y_path, z_path, 'b');
```
