```python
from Dynamics import trivariate_dynamics, trivariate_plot
```

```python
""" 
Functions will all be in the form of Xt+1 = x(Xt, Yt, Zt) - define
function by using Xt, Yt, and Zt

Use numpy for math notation. E.g. np.sqrt(Xt, Yt, Zt)

"""

def f(Xt, Yt, Zt):
    return -0.4*Xt - 0.2*Yt +0.1*Zt

def g(Xt, Yt, Zt):
    return 0.2*Xt - 0.9*Yt + 0.1*Zt

def h(Xt, Yt, Zt):
    return 0.2*Xt - 0.9*Yt + 0.1*Zt
```

```python
x_init = .5
y_init = 1
z_init = .5

x, y, z, x_path, y_path, z_path = trivariate_dynamics(f,
                                                      g,
                                                      h,
                                                      x_init, 
                                                      y_init,
                                                      z_init,
                                                      accuracy = 0.000001,
                                                      length = 10
                                                     )
```
